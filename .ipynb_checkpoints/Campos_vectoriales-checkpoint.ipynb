# Campos vectoriales

Oscilador harmonico simple 
$$k\ddot{x} + kx = 0 $$
En variables de estado 
$$\dot{x} = v$$
$$\dot{v} = - \frac{k}{m}x$$

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
eje = np.arange(-1,1,0.01)
x = eje
y = eje
X,Y = np.meshgrid(x,y)
```

```python
plt.figure(figsize=(10,5))
Bx = Y
By = -10*X
plt.streamplot(X,Y,Bx,By,density = 1,linewidth=1,minlength=0.2)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
```

para el sistema 
$$\begin{bmatrix}  \dot{x} \\ \dot{y} \end{bmatrix}
= \begin{bmatrix}  a & 0 \\ 0 & -1 \end{bmatrix}
= \begin{bmatrix} x \\ y \end{bmatrix}$$

```python
plt.figure(figsize=(20,10))

plt.subplot(2,3,1)
plt.streamplot(X,Y,-2*X,-Y,density = 1,linewidth=1,minlength=0.2)
plt.title('a<-1')

plt.subplot(2,3,2)
plt.streamplot(X,Y,-X,-Y,density = 1,linewidth=1,minlength=0.2)
plt.title('a=-1')

plt.subplot(2,3,3)
plt.streamplot(X,Y,-0.5*X,-Y,density = 1,linewidth=1,minlength=0.2)
plt.title('-1<a<0')

plt.subplot(2,3,4)
plt.streamplot(X,Y,0*X,-Y,density = 1,linewidth=1,minlength=0.2)
plt.title('a=0')

plt.subplot(2,3,5)
plt.streamplot(X,Y,2*X,-Y,density = 1,linewidth=1,minlength=0.2)
plt.title('a>0')
```
