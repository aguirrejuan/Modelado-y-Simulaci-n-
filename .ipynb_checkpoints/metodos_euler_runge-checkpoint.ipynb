<center>MÉTODOS NUMÉRICOS PARA RESOLVER ECUACIONES DIFERENCIALES: <center>Euler y Runge Kutta

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

Método de Euler 

```python
def euler(dx,xo,ti,tf):
    h =0.001
    t = np.arange(ti,tf,h)
    x = np.zeros(len(t))
    x[0]=xo
    for i in range(1,len(t)):
        x[i]= x[i-1] + h*dx(t[i-1],x[i-1])
        
    return t,x
```

Método de Runge Kutta 

```python
def kutta(dx,xo,ti,tf):
    h = 0.001 
    t = np.arange(ti,tf,h)
    x = np.zeros(len(t))
    x[0] = xo
    
    for i in range(1,len(t)):
        k_1 = dx(t[i-1],x[i-1])
        k_2 = dx(t[i-1] + (h/2),x[i-1]+(h/2)*k_1)
        k_3 = dx(t[i-1] + (h/2),x[i-1]+(h/2)*k_2)
        k_4 = dx(t[i-1] + h,x[i-1]+h*k_3)
        x[i] = x[i-1] + (h/6)*(k_1 + 2*k_2 + 2*k_3 + k_4)

    return t,x 
```

## Ejemplo 

$$ \frac{dx}{dt} = sin(2\pi ft)$$

cuya solución es 

$$ x(t) = - \frac{1}{2\pi f}cos(2\pi ft)$$

```python
fs = 10000 
t = np.arange(0,10,1/fs)
def dx(t,x = None):
    f = 1/(2*np.pi)
    return np.sin(t*2*np.pi*f)
```

```python
d_x = dx(t)
```

```python
plt.plot(t,d_x)
plt.xlabel("t")
plt.ylabel("dx/dt")
plt.grid()
```

### Solución usando el método de Euler 

```python
xo = -1
ti,tf = 0,10 
t,x = euler(dx,xo,ti,tf)
plt.plot(t,x)
plt.xlabel("t")
plt.ylabel("dx/dt")
plt.grid()
```

### Solución usando el método de Runge Kutta

```python
xo = -1
ti,tf = 0,10 
t,x = kutta(dx,xo,ti,tf)
plt.plot(t,x)
plt.xlabel("t")
plt.ylabel("dx/dt")
plt.grid()
```
