Encontrar los valores de $x$ para que la función sea igual a $0$

$$f(x) = 0 $$

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
fs = 1000000
x = np.arange(0,2,1/fs)
g = x 
h = np.exp(-x)
f = h - g
```

```python
plt.plot(x,g,x,h,x,f)
plt.xlabel('x')
plt.legend(('g(x)', 'h(x)', 'f(x)'),prop = {'size': 10}, loc='upper right')
plt.grid()
```

La idea es encontrar el punto de intersección 

### Bisection Algorithm 

```python
dx = lambda x: -x**2 +1 
```

```python
t = np.arange(-2,2,0.01)
plt.plot(t,dx(t))
plt.grid()
```

```python
def bisection(dx,x_m,x_p):
    tol = 1e-15
    diff = np.abs(x_m-x_p)
    f_x_p =  dx(x_p)

    while diff>tol: 
        x_next = (x_m + x_p)/2
        f_x_next =  dx(x_next)
        f_x_p =  dx(x_p)
        if f_x_next*f_x_p >0:
            x_p = x_next 
        else:
            x_m = x_next 
        diff = np.abs(x_m-x_p)
    return x_m
```

```python
x_m = -2
x_p = 2
```

```python
bisection(dx,x_m,x_p)
```

### Newton-Raphson Algorithm 

```python
x = np.linspace(-2,5,100)
y = np.exp(-x) - x
plt.plot(x,y)
def f(x): return np.exp(-x)-x

def derivada(fx,x):
    h = 0.001
    return (fx(x+h)-fx(x))/h 

xi=-0.5


def newton(f,xi):
    tol=1e-19
    x_next= xi-(f(xi)/derivada(f,xi))
    diff=np.abs(xi-x_next)
    while tol<diff:
        x_next= xi-(f(xi)/derivada(f,xi))
        diff=np.abs(xi-x_next)
        xi=x_next
    return xi 

xi = newton(f,xi)

plt.plot(xi,f(xi),'ro')
plt.grid()
```
